```python
base_url = 'https://raw.githubusercontent.com/'

list_df_emprestimos = []
for file_name in files_list:
    file_url = base_url + file_name.replace('blob', 'refs/heads')
    df = pd.read_csv(file_url, encoding='utf-8')
    list_df_emprestimos.append(df)
list_df_emprestimos[0].head()
```

```python
df_emprestimos = pd.concat(list_df_emprestimos)
```

```python
df_emprestimos.shape
```

```python
df_exemplares = pd.read_parquet('https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/refs/heads/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
df_exemplares.head()
```

```python
df_exemplares.shape
```

```python
df_exemplares.dropna(inplace=True, how='all')
df_exemplares.shape
```

```python
df_emprestimos.set_index('codigo_barras').join(df_exemplares.set_index('codigo_barras'))
```

```python
df_emprestimos = df_emprestimos.drop_duplicates()
```

```python
df_emprestimos.value_counts()
```

```python
emprestimos_completo = df_emprestimos.merge(df_exemplares)
emprestimos_completo
```

```python
emprestimos_completo['localizacao'].unique()
```

```python
classes = [
    "Generalidades. Ciência e conhecimento.",
    "Filosofia e psicologia.",
    "Religião.",
    "Ciências sociais.",
    "Classe vaga. Provisoriamente não ocupada.",
    "Matemática e ciências naturais.",
    "Ciências aplicadas.",
    "Belas artes.",
    "Linguagem. Língua. Linguística.",
    "Geografia. Biografia. História."
]

classes_dict = {}
for key, classe in enumerate(classes):
    classes_dict[key] = classe

def find_class(localizacao):
    key = int(localizacao/100)
    return classes_dict[key]

classes_dict
```

```python
find_class(630)
```

```python
emprestimos_completo['classe'] = emprestimos_completo['localizacao'].apply(find_class)
```
