```python
base_url = 'https://raw.githubusercontent.com/'

list_df_emprestimos = []
for file_name in files_list:
    file_url = base_url + file_name.replace('blob', 'refs/heads')
    df = pd.read_csv(file_url, encoding='utf-8')
    list_df_emprestimos.append(df)
list_df_emprestimos[0].head()
```

```python
classes = [
    "Generalidades. Ciência e conhecimento.",
    "Filosofia e psicologia.",
    "Religião.",
    "Ciências sociais.",
    "Classe vaga. Provisoriamente não ocupada.",
    "Matemática e ciências naturais.",
    "Ciências aplicadas.",
    "Belas artes.",
    "Linguagem. Língua. Linguística.",
    "Geografia. Biografia. História."
]

classes_dict = {}
for key, classe in enumerate(classes):
    classes_dict[key] = classe

def find_class(localizacao):
    key = int(localizacao/100)
    return classes_dict[key]

classes_dict
```

```python
find_class(630)
```

```python
emprestimos_completo['classe'] = emprestimos_completo['localizacao'].apply(find_class)
```

## Cria um dataframe com a data e quantidade de empréstimos

```python
emprestimos_data = pd.DataFrame(emprestimos_completo['data_emprestimo'].value_counts()).reset_index()
emprestimos_data.columns = ['data','quantidade']
emprestimos_data['ano'] = pd.to_datetime(emprestimos_data['data']).dt.year
emprestimos_data.drop('data', axis=1, inplace=True)
emprestimos_data
```

# Gráfico da quantidade total de exemplares emprestados em cada mês

```python
emprestimos_completo['mes_emprestimo'] = pd.to_datetime(emprestimos_completo['data_emprestimo']).dt.month # strftime("%B")
exemplares_por_mes = emprestimos_completo.groupby('mes_emprestimo')['id_exemplar'].nunique()

exemplares_por_mes_dict = {'mes': exemplares_por_mes.index.tolist(), 'exemplares': exemplares_por_mes.values.tolist()}
exemplares_por_mes = pd.DataFrame(exemplares_por_mes_dict)

with plt.style.context('fivethirtyeight'):
    fig, ax = plt.subplots(figsize=(8,4))
    ax.plot('mes', 'exemplares', data=exemplares_por_mes)
    ax.set_title('Quantidade total de exemplares emprestados em cada mês do ano entre 2010 e 2020')
    ax.set_xlabel('Mês')
    ax.set_ylabel('Exemplares')
    ax.xaxis.set_major_locator(plt.MultipleLocator(1))
    plt.grid(linestyle='--')
    plt.show()
```

# Gráfico da quantidade total de empréstimos em cada hora do dia

```python
emprestimos_completo['hora_emprestimo'] = pd.to_datetime(emprestimos_completo['data_emprestimo']).dt.hour
emprestimos_por_hora = emprestimos_completo.groupby('hora_emprestimo').size()

emprestimos_por_hora_dict = {'hora': emprestimos_por_hora.index.tolist(), 'emprestimos': emprestimos_por_hora.values.tolist()}
emprestimos_por_hora = pd.DataFrame(emprestimos_por_hora_dict)

with plt.style.context('fivethirtyeight'):
    fig, ax = plt.subplots(figsize=(8,4))
    ax.bar('hora', 'emprestimos', data=emprestimos_por_hora)
    ax.set_title('Quantidade total de exemplares emprestados em cada horário')
    ax.set_xlabel('Hora')
    ax.set_ylabel('Empréstimos')
    ax.xaxis.set_major_locator(plt.MultipleLocator(2))
    ax.yaxis.set_major_locator(plt.MultipleLocator(25000))
    plt.grid(linestyle='--')
    plt.show()
```
